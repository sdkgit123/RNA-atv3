```python
#importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#abrindo arquivos e vendo elementos
datavinhored = pd.read_csv('winequality-red.csv', sep=';')
datavinhowhite = pd.read_csv('winequality-white.csv', sep=';')
datavinhored.head()
```

```python
datavinhowhite.head()
```

```python
#numero de elementos em cada arquivo
print(len(datavinhored))
print(len(datavinhowhite))
```

```python
#atributos dos datasets
print(datavinhored.columns)
print(datavinhowhite.columns)
```

```python
#analisando balanceamento de atributos dos datasets
#plotando colunhas vinhored
for i in datavinhored.columns:
    plt.hist(datavinhored[i])
    plt.title(i)
    plt.show()
```

```python
#vinhowhite
for i in datavinhowhite.columns:
    plt.hist(datavinhowhite[i])
    plt.title(i)
    plt.show()
```
